3. [15 poin] Melakukan praproses data. Detail untuk pengerjaan praproses data adalah sebagai berikut.

* Pisahkan bagian atribut prediktor dengan label data menjadi dua variabel terpisah
* Gunakan Simplelmputer untuk menangani missing values, dengan nilai rata-rata dari atribut data sebagai strategi pengisian nilai
* Terapkan StandardScaler untuk mengubah nilai atribut prediktor berjenis numerik
* Terapkan praproses data tambahan (jika dibutuhkan)
* **Tampilkan** statistik deskriptif dari atribut data sebelum dan setelah dilakukan praproses data

```python
from google.colab import drive
drive.mount('/content/drive')

import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler


path = "/content/drive/MyDrive/ML/data uts ml.csv"
df = pd.read_csv(path)

print("=== INFO DATASET AWAL ===")
print("Ukuran data (baris, kolom):", df.shape)
print("Nama kolom:", list(df.columns))
print()

# (a) Pisahkan bagian atribut prediktor dengan label data menjadi dua variabel terpisah
print("=== (a) Memisahkan atribut prediktor (X) dan label (y) ===")
X = df.drop(columns=["HeartDisease"])
y = df["HeartDisease"]

print("Shape X (prediktor):", X.shape)
print("Shape y (label)    :", y.shape)
num_cols = X.select_dtypes(include=["int64", "float64"]).columns.tolist()
cat_cols = X.columns.difference(num_cols).tolist()

print("Kolom numerik   :", num_cols)
print("Kolom kategorik :", cat_cols)
print()

stat_before = X[num_cols].describe().T

# (b) Gunakan Simplelmputer untuk menangani missing values, dengan nilai rata-rata dari atribut data sebagai strategi pengisian nilai
print("=== (b) Imputasi missing value numerik dengan SimpleImputer (mean) ===")

num_imputer = SimpleImputer(strategy="mean")
X_num = X[num_cols]
X_num_imputed = num_imputer.fit_transform(X_num)

X_num_imputed_df = pd.DataFrame(
    X_num_imputed,
    columns=num_cols,
    index=X.index
)

print("Contoh 5 baris kolom numerik setelah imputasi:")
print(X_num_imputed_df.head())
print()

# (c)  Terapkan StandardScaler untuk mengubah nilai atribut prediktor berjenis numerik
print("=== (c) StandardScaler untuk menstandarkan kolom numerik ===")

scaler = StandardScaler()
X_num_scaled = scaler.fit_transform(X_num_imputed_df)

X_num_processed = pd.DataFrame(
    X_num_scaled,
    columns=num_cols,
    index=X.index
)

print("Contoh 5 baris kolom numerik setelah StandardScaler:")
print(X_num_processed.head())
print()


# (d) Terapkan praproses data tambahan (jika dibutuhkan)
print("=== (d) Praproses tambahan: imputasi atribut kategorik (most_frequent) ===")

if cat_cols:
    cat_imputer = SimpleImputer(strategy="most_frequent")
    X_cat = X[cat_cols]
    X_cat_imputed = cat_imputer.fit_transform(X_cat)

    X_cat_processed = pd.DataFrame(
        X_cat_imputed,
        columns=cat_cols,
        index=X.index
    )
    X_processed = pd.concat([X_num_processed, X_cat_processed], axis=1)
else:
    X_processed = X_num_processed.copy()

print("Shape X setelah seluruh praproses:", X_processed.shape)
print("Contoh 5 baris X SETELAH praproses lengkap:")
print(X_processed.head())
print()

# (e) Tampilkan statistik deskriptif dari atribut data sebelum dan setelah dilakukan praproses data
print("=== (e) Statistik deskriptif sebelum & sesudah praproses (kolom numerik) ===")

stat_before_round = stat_before.round(2)
stat_after = X_num_processed.describe().T.round(2)

print("\n--- SEBELUM praproses (data numerik asli) ---")
print(stat_before_round)

print("\n--- SETELAH praproses (setelah imputasi + StandardScaler) ---")
print(stat_after)
```

4. [15 poin] Membangun model klasifikasi berdasarkan dataset yang telah dilakukan praproses data. Detail untuk pengerjaan klasifikasi adalah sebagai berikut.
* Gunakan algoritma pengklasifikasi apapun untuk membangun model klasifikasi. Parameter yang digunakan boleh disesuaikan dengan bebas
* Gunakan metode holdout untuk membagi keseluruhan dataset menjadi data training dan data testing, dengan proporsi 85% data training dan 15% data testing
* Tampilkan confusion matrix dan akurasi dari model klasifikasi

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score

# (a)  Gunakan algoritma pengklasifikasi apapun untuk membangun model klasifikasi. Parameter yang digunakan boleh disesuaikan dengan bebas

print("=== (a) Model klasifikasi yang digunakan ===")
print("Kolom kategorik yang akan di-encode:", cat_cols)
X_model = pd.get_dummies(
    X_processed,
    columns=cat_cols,
    drop_first=True
)

print("Shape fitur SETELAH encoding :", X_model.shape)
model = LogisticRegression(
    max_iter=1000,
    solver="lbfgs"
)

print("Model yang dipakai:")
print(model)
print()

# (b)  Gunakan metode holdout untuk membagi keseluruhan dataset menjadi data training dan data testing, dengan proporsi 85% data training dan 15% data testing

print("=== (b) Pembagian data training dan testing (holdout 85% : 15%) ===")
X_train, X_test, y_train, y_test = train_test_split(
    X_model,
    y,
    test_size=0.15,
    random_state=42,
    stratify=y
)

print("Jumlah data training :", X_train.shape[0])
print("Jumlah data testing  :", X_test.shape[0])
print("Jumlah fitur         :", X_train.shape[1])
print()

# (c)Tampilkan confusion matrix dan akurasi dari model klasifikas
print("=== (c) Training model & evaluasi (confusion matrix + akurasi) ===")
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

cm = confusion_matrix(y_test, y_pred)
cm_df = pd.DataFrame(
    cm,
    index=["Actual 0", "Actual 1"],
    columns=["Pred 0", "Pred 1"]
)

print("Confusion Matrix:")
print(cm_df)

acc = accuracy_score(y_test, y_pred)
print("\nAkurasi model di data testing: {:.4f}".format(acc))
```

5. [15 poin] Membangun model clustering berdasarkan dataset yang telah dilakukan praproses data. Detail untuk pengerjaan clustering adalah sebagai berikut.
* Gunakan algoritma clustering apapun untuk melakukan clustering. Parameter yang digunakan boleh disesuaikan dengan bebas
* Tampilkan SSE dan silhouette score dari model clustering

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
# (a)  Gunakan algoritma clustering apapun untuk melakukan clustering. Parameter yang digunakan boleh disesuaikan dengan bebas
print("=== (a) Model clustering yang digunakan ===")

n_clusters = 2   # boleh kamu ganti: 2, 3, 4, dst
kmeans = KMeans(
    n_clusters=n_clusters,
    random_state=42,
    n_init=10
)

print("Algoritma       : KMeans")
print(f"Jumlah cluster  : {n_clusters}")
print(f"Jumlah data     : {X_model.shape[0]}")
print(f"Jumlah fitur    : {X_model.shape[1]}\n")

kmeans.fit(X_model)
labels = kmeans.labels_  # label cluster untuk tiap data

# (b) Tampilkan SSE dan silhouette score dari model clustering
print("=== (b) SSE dan silhouette score dari model clustering ===")
sse = kmeans.inertia_
sil_score = silhouette_score(X_model, labels)

print(f"Sum of Squared Errors (SSE)      : {sse:.4f}")
print(f"Silhouette Score (k={n_clusters}) : {sil_score:.4f}")
```
